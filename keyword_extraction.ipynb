```python
from transformers import BertTokenizer

# Initialize the tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

def preprocess_text(sentence, max_length=64):
    # Tokenize the sentence
    tokens = tokenizer.tokenize(sentence)

    # Add special tokens ([CLS] and [SEP])
    tokens = ['[CLS]'] + tokens + ['[SEP]']

    # Truncate or pad the tokens to the specified max length
    tokens = tokens[:max_length]
    padding = [0] * (max_length - len(tokens))
    token_ids = tokenizer.convert_tokens_to_ids(tokens) + padding

    # Create attention mask
    attention_mask = [1] * len(tokens) + [0] * len(padding)

    return token_ids, attention_mask

# Example usage
sentence1 = "The cat is sitting on the mat"
sentence2 = "a feline is lounging on the carpet"
token_ids_1, attention_mask_1 = preprocess_text(sentence1)
token_ids_2, attention_mask_2 = preprocess_text(sentence2)

print("Token IDs for sentence 1:", token_ids_1)
print("Attention mask for sentence 1:", attention_mask_1)
print("Token IDs for sentence 2:", token_ids_2)
print("Attention mask for sentence 2:", attention_mask_2)
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from rake_nltk import Rake

# Create a Rake instance
r = Rake()

# Text from which keywords will be extracted
text = "RAKE (Rapid Automatic Keyword Extraction) is a keyword extraction algorithm that automatically identifies relevant keywords and phrases in a text document."

# Extract keywords from the text
r.extract_keywords_from_text(text)

# Get the ranked keywords
keywords = r.get_ranked_phrases_with_scores()

# Print the extracted keywords and their scores
for score, kw in keywords:
    print("Keyword:", kw, "Score:", score)
```

```python
!pip3 install spacy
```

```python
import spacy
from collections import Counter
from string import punctuation
nlp = spacy.load("en_core_web_sm")
def get_hotwords(text):
    result = []
    pos_tag = ['PROPN', 'ADJ', 'NOUN']
    doc = nlp(text.lower())
    for token in doc:
        if(token.text in nlp.Defaults.stop_words or token.text in punctuation):
            continue
        if(token.pos_ in pos_tag):
            result.append(token.text)
    return result
text = """
spaCy is an open-source natural language processing library for Python.
"""
output = set(get_hotwords(text))
most_common_list = Counter(output).most_common(10)
for item in most_common_list:
  print(item[0])
```

```python

from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
import spacy

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Sample text input
text = """A word cloud is a visualization technique used to represent text data, where the size of each word corresponds to its frequency or importance within the text. In a word cloud, words are typically arranged randomly or in a visually appealing manner, with more frequent words typically displayed in larger font sizes and less frequent words displayed in smaller font sizes."""

# Step 1: Tokenization
doc = nlp(text)

# Step 2: Filtering (Optional)
# Optionally, you can remove stop words or apply other preprocessing steps here

# Step 3: Word Frequency Calculation
word_freq = Counter(token.text for token in doc)

# Step 4: Generate Word Cloud
wordcloud = WordCloud(width=800, height=400, background_color="white").generate_from_frequencies(word_freq)

# Step 5: Display the Word Cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
